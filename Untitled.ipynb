```python
import pandas as pd

df = pd.read_csv('train (1).csv')
df
```

```python
df.groupby(['Sex', 'Pclass'], as_index = False)\
.agg({'Fare':['sum','mean'],'Name':['count']})\
.round(2)
```

**Выводы:** 1.большинство биллетов было продано на 3й класс(у мужчин третий класс пользовался особой популярностью). 2. Мужчин во всех классах заметно больше.3. Средняя стоимость биллета для женщин была больше чем для мужчин, особенно заметно в первом классе почти в два раза.
